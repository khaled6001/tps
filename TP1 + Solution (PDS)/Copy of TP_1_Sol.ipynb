```python
#@title RectArea function.
def RectArea(w,h):
    return w*h
print(RectArea(5,10))
print(RectArea(5,"b")) # python is multiplying the values even if they're not numeric (int & float), the result is incorrect.
```

```python
#@title RectArea function with errors handling.
def IntOrFloat(x): # IntOrFloat allows us to check the parameter x if it's a numeric value or not (int or float), it returns true if x is int or float.
    return isinstance(x, int) or isinstance(x, float)

def RectArea(w,h):
    if(not IntOrFloat(w) or not IntOrFloat(h)): # here we check w & h, both must be numeric value.
        print("only int and float are accepted.")
        return None
    return w*h

print(RectArea(5,10))
print(RectArea(5,"5"))
```

```python
#@title Complicity/Execution time
import sys
import numpy as np

# Create two python arrays py_arr1 and py_arr2 with length of 1000000 (their values are from 0 to 999999).
py_arr_1 = list(range(1000000))
py_arr_2 = list(range(1000000))

# Create two numpy arrays np_arr1 and np_arr2 with length of 1000000 (their values are from 0 to 999999).
np_arr_1 = np.arange(1000000)
np_arr_2 = np.arange(1000000)

%time [x+y for (x,y) in zip(py_arr_1, py_arr_2)] # calculation of the sum of py_arr1 &  py_arr2 using a FOR loop.
%time np_arr_1 + np_arr_2 # calculation the sum of np_arr_1 & np_arr_2 using + operation (addition).

# we can notice that the excution time is faster x50 times with numpy.
```

```python
#@title Rounding
import numpy as np
np_arr = np.random.randn(6,8) # create an array np_arr of random values with dimensions of 6x8.
print(np_arr)
print(np_arr.round(2)) # Round np_array to two decimal places. (eg. 0.12345..9 â†’ 0.12)``
print(np_arr.round(0)) # Using the previous array create np_arr_int_rounded (with the rounded integer values).
np_arr_int = np.array(np_arr, dtype= np.int32) # Using the previous array create np_arr_int (without the rounding the integer values).
print(np_arr_int)
```
