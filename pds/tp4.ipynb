```python
import pandas as pd
import numpy as np

space = ["2D", "2D", "2D", "3D", "2D", "2D", "2D", "3D", "2D", "3D", "2D"]
shape = ["Triangle", "Triangle", "Triangle", "Cube", "Rectangle", "Triangle", "Circle", "Sphere", "Circle", "Cube", "Rectangle"]
space = pd.Series(space)
shape = pd.Series(shape)

index = {"index": [space, shape]}
data1 = [15.24, 18.27, 16.08, 2.95, 9.68, 20.87, 20.15, 4.00, 12.26, 8.41, 7.61]
data2 = [18.67, 20.91, 9.66, 13.56, 12.2, 13.86, np.nan, np.nan, np.nan, 21.3, 10.69]
data3 = [np.nan, np.nan, np.nan, 19.21, np.nan, np.nan, np.nan, np.nan, np.nan, 8.09, np.nan]

df = {
    "data1": data1,
    "data2": data2,
    "data3": data3,
}

df = pd.DataFrame(df, columns=["data1", "data2", "data3"])
multi_index = pd.MultiIndex.from_arrays(index["index"], names=["space", "shape"])
df.set_index(multi_index, inplace=True)

# Rename the multi-level index to "index"
df = df.rename_axis(index="index")



df
```
